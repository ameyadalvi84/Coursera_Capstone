##### Selecting the data as per our requirement
The reason for choosing the fields mentioned below can be found in the notebook - https://github.com/ameyadalvi84/Coursera_Capstone/blob/master/Capstone%20Project.ipynb

```python
df_intermediate = data[['ADDRTYPE','COLLISIONTYPE','PERSONCOUNT','VEHCOUNT','WEATHER','ROADCOND','LIGHTCOND','SEVERITYCODE']]
df_intermediate.head(10)
```

```python
one_hot_columns = ['ADDRTYPE','COLLISIONTYPE','WEATHER','ROADCOND','LIGHTCOND']
#df_intermediate.isnull().sum()
```

We are going to drop rows which have Null records.

```python
df_intermediate.dropna(inplace=True)
df_intermediate.isnull().sum()
```

Creating Dummy colums for categorical variables

```python
df_final=pd.get_dummies(df_intermediate, columns=one_hot_columns, drop_first=True)
df_final.info()
```

Checking to see if there is any correlation between various parameters

```python
df_final.corr()
```

Splitting the data into train and test data

```python
from sklearn.model_selection import train_test_split
from sklearn.utils import resample
```

```python
train,test = train_test_split(df_final,test_size = 0.15,random_state = 140)
```

```python
print(train.shape)
print(test.shape)
```

Need to balance the skewed data, by resampling the Severity =1 dataset.

```python
bal = train['SEVERITYCODE'].value_counts()
bal[2]
```

```python
train_bal_1 = resample(train[train['SEVERITYCODE']==1],
                  replace=False,
                      n_samples = bal[2],
                      random_state=140)
train_bal_1.head()
```

```python
train_bal_2 = pd.DataFrame(train[train['SEVERITYCODE']==2])
train_bal_2.shape
```

Creating the dataset with equal count of data for Severity 1 and Severity 2.

```python
downsample_train = pd.concat([train_bal_1,train_bal_2])
downsample_train.shape
```

Normalizing the data for Person Count and Vehical Count

```python
from sklearn.preprocessing import StandardScaler

scale1 = StandardScaler(copy=False)
scale2 = StandardScaler(copy=False)
scale1.fit(downsample_train[['PERSONCOUNT']])
downsample_train['PerCount'] = scale1.transform(downsample_train[['PERSONCOUNT']])
scale2.fit(downsample_train[['VEHCOUNT']])
downsample_train['VehCount'] = scale2.transform(downsample_train[['VEHCOUNT']])
downsample_train.head()
```

```python
X_train = downsample_train.drop(['SEVERITYCODE','PERSONCOUNT','VEHCOUNT'],axis=1)
y_train = downsample_train['SEVERITYCODE']
```

```python
test['PerCount'] = scale1.transform(test[['PERSONCOUNT']])
test['VehCount'] = scale2.transform(test[['VEHCOUNT']])
#test.head()
X_test = test.drop(['SEVERITYCODE','PERSONCOUNT','VEHCOUNT'],axis=1)
y_test = test['SEVERITYCODE']
```

```python
assert X_train.shape[0] == y_train.shape[0]
assert X_test.shape[0] == y_test.shape[0]
```

### Model1 - Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
clf = LogisticRegression()
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

Checking the f1 score for the model

```python
print(classification_report(y_test,y_pred))
```

```python
parameters = {'C':(1,1e5,1e2),
    'solver':('newton-cg', 'lbfgs', 'sag')}
grid_clf = GridSearchCV(LogisticRegression(penalty = 'l2',n_jobs=-1),param_grid=parameters)
```
